```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df=pd.read_csv(r"C:\Users\panka\Machine Learning\Fake News\data.csv")
```

```python
df
```

```python
df_main=df[['Headline','Label','Body']]
```

```python
df_main
```

```python
df_new=df_main.copy()
```

```python
df_new['Body'].fillna('no info', inplace = True)
```

```python
df_new.isnull().sum()
```

```python
df_x=df_new['Body']
df_y=df_new['Label']
```

```python
x_train, x_test, y_train, y_test=train_test_split(df_x,df_y, test_size=0.2, random_state=12)
```

```python
tv1=TfidfVectorizer(min_df=1,stop_words='english')
```

```python
new_xtrain=tv1.fit_transform(x_train)
```

```python
a=new_xtrain.toarray()
```

```python
tv1.get_feature_names()
```

```python
a
```

```python
tv1.inverse_transform(a[0])
```

# USING SVM

```python
from sklearn.svm import SVC
model_svm=SVC(kernel="linear")
#y_train=y_train.astype('int')
model_svm.fit(new_xtrain,y_train)
```

```python
new_xtest=tv1.transform(x_test)
```

```python
pred2=model_svm.predict(new_xtest)
```

```python
y_test1=y_test.astype('int')
```

```python
from sklearn.metrics import accuracy_score as ac
ac(pred2,y_test1)
```

# USING KNN

```python
from sklearn.neighbors import KNeighborsClassifier 
knn = KNeighborsClassifier(n_neighbors=4) 
knn.fit(new_xtrain,y_train)
```

```python
pred1=knn.predict(new_xtest)
```

```python
from sklearn.metrics import accuracy_score as ac
ac(pred1,y_test1)
```

# USING RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(new_xtrain,y_train)
```

```python
pred=rf.predict(new_xtest)
```

```python
from sklearn.metrics import accuracy_score as ac
ac(pred,y_test1)
```
